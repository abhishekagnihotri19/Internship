Rank
B) Name
C) Artist
D) Upload date
E) Views 

```python
driver1=webdriver.Chrome()
try:
    driver1.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos ')
    table_1=driver1.find_element(By.XPATH,'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[2]' )
    table_row=table_1.find_elements(By.TAG_NAME,'tr')
    for row in table_row:
        columns=row.find_elements(By.TAG_NAME,'td')

        if len(columns)>=5:
            rank=columns[0].text
            video=columns[1].text
            artist=columns[2].text
            viewer=columns[3].text
            published=columns[4].text

            print(f'Rank:{rank}, Video-name{video},Artist:{artist},NumberOfViewers:{viewer},Date_Launch:{published}')
except Exception as e:
    print(f'There is one exception occur:{e}')
        

```

Note: - From bcci.tv home page you have reach to the international fixture page through code.

# Q7. Scrape the details most watched tv series of all time from imdb.com.
Url = https://www.imdb.com/list/ls095964455/ You have
to find the following details:
A) Name
B) Year span
C) Genre
D) Run time
E) Ratings
F) Votes 

```python
driver7 = webdriver.Chrome()

try:
    driver7.get('https://www.imdb.com/list/ls095964455/')
    name = []
    year = []
    genre = []
    run_time = []
    ratings = []
    votes = []

    names = driver7.find_elements(By.XPATH, '//h3[@class="lister-item-header"]')
    for i in names:
        only_name = i.text
        name.append(only_name)

    years = driver7.find_elements(By.XPATH, '//span[@class="lister-item-year text-muted unbold"]')
    for j in years:
        only_years = j.text
        year.append(only_years)

    genres = driver7.find_elements(By.XPATH, '//span[@class="genre"]')
    for k in genres:
        only_genre = k.text
        genre.append(only_genre)

    runtym = driver7.find_elements(By.XPATH, '//span[@class="runtime"]')
    for m in runtym:
        only_runtym = m.text
        run_time.append(only_runtym)

    rates = driver7.find_elements(By.XPATH, '//div[@class="ipl-rating-star small"]/span[2]')
    for r in rates:
        only_rated = r.text
        ratings.append(only_rated)

    vote_elements = driver7.find_elements(By.CSS_SELECTOR, 'span[name="nv"]')
    for v in vote_elements:
        only_votes = v.text
        votes.append(only_votes)

    #print(len(name), len(year), len(genre), len(run_time), len(ratings), len(votes))
    
    df7=pd.DataFrame({'SONG-NAME':name,'YEAR-NAME':year,'RUN-TIME':run_time,'RATE':ratings,'VOTE':votes})
    print(df7)

except Exception as e:
    print(f'There is an Exception: {e}')
finally:
    driver7.quit()
```
