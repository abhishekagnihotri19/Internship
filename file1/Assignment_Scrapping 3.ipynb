```python
def search_all(myInput):
    driver1=webdriver.Chrome()
    try:
        driver1.get('https://www.amazon.in/')
        search_input=driver1.find_element(By.ID,"twotabsearchtextbox")
        search_input.send_keys(myInput)
        #search_input.send_keys(Keys.RETURN)

        
        search_lap_start=driver1.find_element(By.ID,"nav-search-submit-button")
        search_lap_start.click()
        #driver1.implicitly_wait(5)

        
        time.sleep(8)
        products_name=[]
        listed_products=driver1.find_elements(By.XPATH,'//div[@class="a-section a-spacing-none a-spacing-top-small s-title-instructions-style"]/h2')

        for p in listed_products:
            name_products=p.text
            products_name.append(name_products)
            #print(products_name)
            if len(products_name)>=50:
                break

        
        product_Price=[]
        product_prices=driver1.find_elements(By.XPATH,'//span[@class="a-price-whole"]')
        
        for q in product_prices:
            price_guitar=q.text
            product_Price.append(price_guitar)
            if len(product_Price)>=50:
                break
            #print(price_guitar)
            
        exp_delivery=[]
        del_dates=driver1.find_elements(By.XPATH,'//span[@class="a-color-base a-text-bold"]')
        
        for d in del_dates:
            deliveryDates=d.text
            exp_delivery.append(deliveryDates)
            if len(exp_delivery)>=50:
                break
           # print(exp_delivery)
        
        prod_url=[]
        all_url=driver1.find_elements(By.XPATH,'//h2[@class="a-size-mini a-spacing-none a-color-base s-line-clamp-4"]/a')
        for u in all_url:
            my_urls=u.text
            prod_url.append(my_urls)
            if len(prod_url)>=50:
                break
         
        #print(len(products_name))
        #print(len(product_Price))
        #print(len(exp_delivery))
        #print(len(prod_url))

        guitars=pd.DataFrame({"Brand-Name":products_name,"Price":product_Price,"Delivery":exp_delivery,"Product-url":prod_url})
        print(guitars)
        
        
    
    except  Exception as e:
         print(f"An error occurred: {e}")
    finally:
        # Close the browser window
        driver1.quit()
if __name__ == "__main__":
    user_input = input("Enter the product to search on Amazon: ")
    search_all(user_input)
```

```python
import os
import time
import base64
from io import BytesIO
from PIL import Image
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import requests

def cake_googles(user_cake, cake_folder):
    driver3 = webdriver.Chrome()
    try:
        driver3.get('https://images.google.com/')
        cake_only = driver3.find_element(By.ID, 'APjFqb')
        cake_only.send_keys(user_cake)
        cake_only.send_keys(Keys.RETURN)
        time.sleep(10)
        
        cake_img = driver3.find_elements(By.TAG_NAME, 'img')
        
        if not os.path.exists(cake_folder):
            os.makedirs(cake_folder)
        
        for c, cakes_photos in enumerate(cake_img):
            cake_url = cakes_photos.get_attribute('src')
            
            if cake_url and cake_url.startswith('data:image'):
                base64_removal = cake_url.split(',')[1]
                final_cakes_img = base64.b64decode(base64_removal)
                image = Image.open(BytesIO(final_cakes_img))
                image.save(os.path.join(cake_folder, f'image{c}.png'))

            elif cake_url and cake_url.startswith(('http:', 'https:')):
                # If it's a valid URL, use requests to download the image
                response = requests.get(cake_url)
                image = Image.open(BytesIO(response.content))
                image.save(os.path.join(cake_folder, f'image{c}.png'))

            else:
                continue
    
    except Exception as e:
        print(f'An error occurred: {e}')
    
    finally:
        # Close the browser window
        driver3.quit()

if __name__ == "__main__":
    cake_only = input("Enter your cake name: ")
    cake_folder = "downloaded_cakes_images"
    cake_googles(cake_only, cake_folder)
```
